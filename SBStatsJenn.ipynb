```python
# Import Numpy for calculations and matplotlib for charting
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Import our data into pandas from CSV file
reg_season = 'superbowl reg season stats.csv'
reg_season_df = pd.read_csv(reg_season)

reg_season_df
```

```python
# Get a list of all columns for easy reference
reg_season_df.columns
```

```python
#Drop the last row of data
reg_season_df = reg_season_df.drop([10], axis=0)
reg_season_df
```

```python
#Cleaning up the data and changing float to integers
reg_season_df["Date/Season"] = reg_season_df["Date/Season"].astype(int)
reg_season_df["Score winning"] = reg_season_df['Score winning'].astype(int)
reg_season_df["Score Losing"] = reg_season_df["Score Losing"].astype(int)
reg_season_df["points scored winner"] = reg_season_df["points scored winner"].astype(int)
reg_season_df["points allowed winner"] = reg_season_df["points allowed winner"].astype(int)
reg_season_df["points scored loser"] = reg_season_df["points scored loser"].astype(int)
reg_season_df["points allowed loser"] = reg_season_df["points allowed loser"].astype(int)

reg_season_df
```

```python
#Plot Bar graph for each year for win loss score
labels = reg_season_df['Date/Season']
winner_score = reg_season_df["Score winning"]
loser_score = reg_season_df["Score Losing"]

# the label locations
x = np.arange(len(labels))  

# the width of the bars
width = 0.35  

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, winner_score, width, label='Winner')
rects2 = ax.bar(x + width/2, loser_score, width, label='Loser')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Final Score')
ax.set_title('Final Scores per Last 10 SuperBowls')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 2),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()

```

```python
#Get average calculation for the winners and losers
ave_winner = reg_season_df["Score winning"].mean()
ave_loser = reg_season_df["Score Losing"].mean()
ave_winstd = reg_season_df["Score winning"].std()
ave_losstd = reg_season_df["Score Losing"].std()


#Get definitions 
 
players = ["Winner", "Loser"]
scores =[ave_winner, ave_loser]
means = [ave_winner, ave_loser]
error = [ave_winstd, ave_losstd]

# set x_axis and locations
x_axis = np.arange(0, len(players))
tick_locations = []
for x in x_axis:
    tick_locations.append(x)

#Build the plot with error bars
fig, ax = plt.subplots()
ax.bar(x_axis, means,
       yerr=error,
       align='center',
       alpha=0.5,
       ecolor='black',
       capsize=10)
ax.set_ylabel('Score')
ax.set_xticks(x_axis)
ax.set_xticklabels(players)
ax.set_title('Average Winning Score vs Losing Score')
ax.yaxis.grid(True)

# Save the figure and show
plt.tight_layout()
plt.savefig('bar_plot_with_error_bars.png')
plt.show()
```
