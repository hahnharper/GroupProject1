```python
import pandas as pd
import numpy as np
import requests
import gmaps
import json
import pprint
from citipy import citipy
from config import g_key

# Access maps with unique API key
gmaps.configure(g_key)
```

```python
#Import csv

file = "Resources/super_bowl_data.csv"
sb_csv = pd.read_csv(file)
sb_csv.head(50)
```

```python
cities = pd.DataFrame(sb_csv['City'])
cities = cities.drop_duplicates(keep='first')
cities
```

```python
#Get coordinates for each city

# for itm in locations['results']:
#     print(itm["geometry"]["location"]["lat"])
#     print(itm["geometry"]["location"]["lng"])
    

# Target city
target_city = list(cities)
print(target_city)

city_lat_lng = []

for _,city in cities.iterrows():
    # Build the endpoint URL
    target_url = ('https://maps.googleapis.com/maps/api/geocode/json?'
    'address={0}&key={1}').format(city, g_key)
    print(city['City'])

# Run a request to endpoint and convert result to json
    locations = requests.get(target_url).json()
#     print(json.dumps(locations, indent=4, sort_keys=True))

    lat = locations['results'][0]['geometry']['location']['lat']
    lng = locations['results'][0]['geometry']['location']['lng']

    city_dict = {
        "City":city['City'],
        "Lat":lat,
        "Lng":lng
    }
    
    city_lat_lng.append(city_dict)


data = pd.DataFrame(city_lat_lng)
data.head(40)       
```

```python
# Access maps with unique API key
gmaps.configure(api_key=g_key)
```
